# Usando MatplotLib, Seaborn, Numpy e Pandas na Criacao de Graficos Estatisticos

```python

np.random.seed(42)
n = 1000
pct_smokers = 0.2

flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

dados = pd.DataFrame({'altura': altura, 'peso': peso,
                      'flag_fumante': flag_fumante})

dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'No Fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
sea.set(style = 'ticks')

sea.lmplot(x = 'altura',
           y = 'peso',
           data = dados,
           hue = 'flag_fumante',
           palette = ['tab:blue', 'tab:orange'],
           height = 7)

plt.title('Relacao entre Altura e Peso de Fumantes e Nao Fumantes')

sea.despine()

plt.show()
```
